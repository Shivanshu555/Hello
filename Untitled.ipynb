```python
import re



convert={'¹':1,'²':2,'³':3,'⁴':4,'⁵':5,'⁶':6,'⁷':7,'⁸':8,'⁹':9,'⁰':0,'¹⁰':10,'¹¹':11,'¹²':12,'¹³':13,'¹⁴':14}



def sup(num):

    sup_conv={'1':'¹','2':'²',12'3':'³','4':'⁴','5':'⁵','6':'⁶','7':'⁷','8':'⁸','9':'⁹','0':'⁰'}

    num=str(num)

    for i in num:

        num=num.replace(i,sup_conv[i])

    return num



def arranger(config):



	orbitals=re.findall(r'[1-7][s|p|d|f][\D]+',config)

	order=lambda val: val[0:1]

	orbitals.sort(key=order)

	return ''.join(orbitals)





def e_config(electrons):

    ''' This functions gives electronic configuration on the basis of aufbau's priciple

Note: 1) Exceptions do exist

      2) Electronic Configuration for electrons>118 can be calculated by this method

    '''

  

    if electrons>118:

        raise ValueError(' value greater than 118')

    

    

    orbital_order='''1s>2s>2p>3s>3p>4s>3d>4p>5s>4d>5p>6s>4f>5d>6p>7s>5f>6d>7p'''



    config=''



    if electrons in exceptions:

        return exceptions[electrons]



    for orbital in orbital_order.split('>'):

        subshell=orbital[1:]

        config+=orbital

        

        

        if subshell=='s':



            if (electrons-2)>=0:

                config+='²'

                electrons-=2

            else:

                config+=sup(electrons)

                break

            

        elif subshell=='p':

            if (electrons-6)>=0:

                config+='⁶'

                electrons-=6

            else:

                config+=sup(electrons)

                break

        elif subshell=='d':



            if (electrons-10)>=0:

                config+='¹⁰'

                electrons-=10

            else:

                config+=sup(electrons)

                break



        elif subshell=='f':



            if (electrons-14)>=0:

                config+='¹⁴'

                electrons-=14

            else:

                config+=sup(electrons)

                break



        if electrons==0:

            break

    return config



exceptions={

    24:'1s²2s²2p⁶3s²3p⁶4s¹3d⁵',

    29:'1s²2s²2p⁶3s²3p⁶4s¹3d¹⁰',

    41:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s¹4d⁴',

    42:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s¹4d⁵',

    44:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s¹4d⁷',

    45:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s¹4d⁸',

    46:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶4d¹⁰',

    47:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s¹4d¹⁰',

    

    57:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f¹5d¹',

    

    58:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f²',

    

    64:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f⁷5d¹',

    

    78:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s¹4f¹⁴5d⁹',

    79:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s¹4f¹⁴5d¹⁰',

    

    111:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f¹⁴5d¹⁰6p⁶7s¹5f¹⁴6d¹⁰',

    

    89:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f¹⁴5d¹⁰6p⁶7s²6d¹',

    90:'1s²2s²2p⁶3s²3p⁶4s²3d¹⁰4p⁶5s²4d¹⁰5p⁶6s²4f¹⁴5d¹⁰6p⁶7s²6d²',

    

    91:'1s²2s²2p⁶3s²3p⁶3d¹⁰4s²4p⁶4d¹⁰4f¹⁴5s²5p⁶5d¹⁰5f²6s²6p⁶7s²6d¹',

    92:'1s²2s²2p⁶3s²3p⁶3d¹⁰4s²4p⁶4d¹⁰4f¹⁴5s²5p⁶5d¹⁰5f³6s²6p⁶7s²6d¹',

    93:'1s²2s²2p⁶3s²3p⁶3d¹⁰4s²4p⁶4d¹⁰4f¹⁴5s²5p⁶5d¹⁰5f⁴6s²6p⁶7s²6d¹',



    96:'1s²2s²2p⁶3s²3p⁶3d¹⁰4s²4p⁶4d¹⁰4f¹⁴5s²5p⁶5d¹⁰5f⁷6s²6p⁶7s²6d¹',

    97:'1s²2s²2p⁶3s²3p⁶3d¹⁰4s²4p⁶4d¹⁰4f¹⁴5s²5p⁶5d¹⁰5f⁸6s²6p⁶7s²6d¹',

    }        



while True:

    atomic_num=int(input('\nEnter atomic number of element: '))



    config=e_config(atomic_num)

    print('\nAccording to energy level:\n')

    print(config)

    print('\nArranged:\n')

    print(arranger(config))





##for i in range(1,118):

##	config=arranger(e_config(i))

##	orbitals=re.findall(r'[1-7][s|p|d|f][\D]+',config)

##	num=[convert[i[2:]] for i in orbitals]

##	print(num)

    
```

```python
# Stopwatch

from matplotlib import animation

import matplotlib.pyplot as plt

import time



# Creating Figure

fig = plt.figure()



# Removing frame and axis ticks

ax  = plt.axes(frameon=False)

ax.axes.get_xaxis().set_visible(False)

ax.axes.get_yaxis().set_visible(False)



# For maximizing the graph window

mng = plt.get_current_fig_manager()

mng.full_screen_toggle()



# Our time text

time_text = ax.text(0.1, 0.5, '', fontsize=120)



def t(n):

    ''' Gives hours:minutes:seconds format from 'n' seconds'''

    

    time.sleep(1)

    tim=n

    

    hr=int(str(tim/3600).split('.')[0])

    tim-=hr*3600

    



    mi=int(str(tim/60).split('.')[0])

    tim-=mi*60



    sec=int(round(tim,0))



    if len(str(hr))==1:

        hr='0'+str(hr)

        

    if len(str(mi))==1:

        mi='0'+str(mi)

        

    if len(str(sec))==1:

        sec='0'+str(sec)



    return '{0}:{1}:{2}'.format(hr,mi,sec)





def run_animation():

    '''Runs the animation'''



    # Stopping and Starting of Stopwatch

    anim_running = True

    def onClick(event):

        nonlocal anim_running

        if anim_running:

            anim.event_source.stop()

            anim_running = False

        else:

            anim.event_source.start()

            anim_running = True



    # Updating Stopwatch time

    def updatefig(num):

        time=t(num)

        time_text.set_text(time)

        return time_text,

    

    fig.canvas.mpl_connect('button_press_event', onClick)



    anim = animation.FuncAnimation(fig, updatefig)

    plt.show()



run_animation()

```
